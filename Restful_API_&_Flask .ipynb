9. How do you handle errors in Flask APIs?
- Handling errors effectively in Flask APIs is crucial for building robust and user-friendly applications. Flask provides several mechanisms to manage errors gracefully, ensuring that clients receive meaningful feedback and that developers can debug issues efficiently.
- Use Custom Error Handlers: Define custom error handlers using the @app.errorhandler decorator to return structured JSON responses. This ensures consistency and clarity in error messages.

- Create Custom Exception Classes: Implement custom exception classes for application-specific errors. This allows for more granular control over error handling and response generation.

21. How would you secure a Flask API?
- Securing a Flask API involves multiple layers and best practices to protect it from unauthorized access, data breaches, and attacks.
- Use HTTPS

  Always serve your API over HTTPS to encrypt data in transit.

  Use SSL/TLS certificates.

- Authentication

  Implement token-based authentication like JWT (JSON Web Tokens) or OAuth2.

  Use Flask extensions like Flask-JWT-Extended or Flask-OAuthlib.

  Require clients to send tokens in headers (Authorization: Bearer <token>).

- Authorization

  Define roles and permissions to restrict access to resources.

 Verify user permissions before processing requests.

- Input Validation & Sanitization

 Validate incoming data to prevent injection attacks (SQL injection, XSS).

 Use libraries like marshmallow for schema validation.

- Rate Limiting

 Protect against brute force and DDoS attacks.

 Use Flask-Limiter to limit the number of requests per IP or user.

- Use Secure Headers

 Set HTTP headers like Content-Security-Policy, X-Content-Type-Options, and Strict-Transport-Security.

 Use Flask-Talisman extension to simplify setting security headers.

- Error Handling

 Avoid exposing sensitive information in error messages.

 Customize error responses without revealing stack traces.

- Keep Dependencies Updated

 Regularly update Flask and third-party libraries to patch security vulnerabilities.

- CORS Configuration

 Use Flask-CORS to configure Cross-Origin Resource Sharing securely.

 Allow only trusted origins.

```python
#2. How do you serve static files like images or CSS in Flask?
from flask import Flask, url_for, render_template_string

app = Flask(__name__)

@app.route('/')
def home():
    # Simple HTML referencing static CSS and image files
    html = '''
    <html>
      <head>
        <link rel="stylesheet" type="text/css" href="{{ url_for('static', filename='css/style.css') }}">
      </head>
      <body>
        <h1>Welcome to Flask Static Files!</h1>
        <img src="{{ url_for('static', filename='images/logo.png') }}" alt="Logo" width="200">
      </body>
    </html>
    '''
    return render_template_string(html)

if __name__ == '__main__':
    app.run(debug=True)
```

```python
#3. How do you define different routes with different HTTP methods in Flask?
from flask import Flask, request
app = Flask(__name__)
@app.route('/hello', methods=['GET'])
def hello():
    return "Hello, GET request!"
@app.route('/submit', methods=['POST'])
def submit():
    data = request.form.get('data', 'No data sent')
    return f"Received POST data: {data}"
@app.route('/both', methods=['GET', 'POST'])
def both_methods():
    if request.method == 'POST':
        return "This is a POST request"
    else:
        return "This is a GET request"
if __name__ == '__main__':
    app.run(debug=True)
```

```python
#5. How can you generate URLs for routes in Flask using url_for?
from flask import Flask, url_for
app = Flask(__name__)
@app.route('/')
def home():
    return "Home page"
@app.route('/user/<username>')
def profile(username):
    return f"Profile page of {username}"
@app.route('/show_urls')
def show_urls():
    home_url = url_for('home')
    profile_url = url_for('profile', username='alice')
    return f"Home URL: {home_url} <br> Profile URL: {profile_url}"
if __name__ == '__main__':
    app.run(debug=True)
```

```python
#6. How do you handle forms in Flask?
from flask import Flask, render_template, request
app = Flask(__name__)
@app.route('/form', methods=['GET', 'POST'])
def form():
    if request.method == 'POST':
        name = request.form.get('name')
        email = request.form.get('email')
        return f"Received: Name = {name}, Email = {email}"
    return render_template('form.html')
if __name__ == '__main__':
    app.run(debug=True)
```

```python
#7. How can you validate form data in Flask?
from flask import Flask, request, render_template_string
app = Flask(__name__)
@app.route('/submit', methods=['GET', 'POST'])
def submit():
    error = None
    if request.method == 'POST':
        name = request.form.get('name', '').strip()
        if not name:
            error = "Name is required."
        elif len(name) < 3:
            error = "Name must be at least 3 characters."
        else:
            return f"Hello, {name}!"
    html = '''
    <form method="POST">
        Name: <input type="text" name="name" />
        <input type="submit" value="Submit" />
    </form>
    {% if error %}
    <p style="color:red;">{{ error }}</p>
    {% endif %}
    '''
    return render_template_string(html, error=error)
if __name__ == '__main__':
    app.run(debug=True)
```

```python
#8. How do you manage sessions in Flask?
from flask import Flask, session, redirect, url_for, request
app = Flask(__name__)
app.secret_key = 'secret!'
@app.route('/login', methods=['GET', 'POST'])
def login():
    if request.method == 'POST':
        username = request.form['username']
        session['username'] = username
        return redirect(url_for('profile'))
    return '''
        <form method="post">
            Username: <input name="username" />
            <input type="submit" />
        </form>
    '''
@app.route('/profile')
def profile():
    if 'username' in session:
        return f"Logged in as {session['username']}"
    return "You are not logged in"
@app.route('/logout')
def logout():
    session.pop('username', None)
    return "Logged out"
if __name__ == '__main__':
    app.run(debug=True)
```

```python
#12. How do you define a custom Jinja filter in Flask?
from flask import Flask, render_template_string
app = Flask(__name__)
def reverse_string(s):
    return s[::-1]
app.jinja_env.filters['reverse'] = reverse_string
@app.route('/')
def home():
    return render_template_string('Original: {{ text }} <br> Reversed: {{ text | reverse }}', text="hello")
if __name__ == '__main__':
    app.run(debug=True)
```
